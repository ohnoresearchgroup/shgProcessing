```python
name = path + "CVpH4pt6.csv"


shg_pH4pt6 = SHGfromCCD(name)

#indices to the below function should be chosen to produce a 1 nm range of background
#on either side of the peak (shown in the orange circles)
shg_pH4pt6.backgroundSubtract([514,518])

shg_pH4pt6.plotIntensities()

(shg_pH4pt6_mean,shg_pH4pt6_std) = shg_pH4pt6.getMean()
```
